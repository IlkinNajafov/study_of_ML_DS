# Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Preparing the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(15, 10)) # width and height

# Plot the data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax1.plot(x, x-2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(gun_prices.keys(), gun_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# Subplot option 2
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(gun_prices.keys(), gun_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

### Which one should you use?
### - When plotting something quickly, okay to use
### pyplot method.
### - When plotting something advanced, use the 00
### method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10, 5));
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data

scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add horizontal line

ax.axhline(over_50.chol.mean(), linestyle='--')
```

```python
over_50["target"].values
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x=heart_disease.age,
          y=heart_disease.chol,
          c=heart_disease.target)

ax.set(title='Heart Disease',
      xlabel='age',
      ylabel='chol')
ax.set_ylim(100, 400)
ax.set_xlim(30, 70)

ax.legend(*scatter.legend_elements(), title='target')

ax.axhline(over_50.chol.mean(),
            linestyle='--')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(x=over_50.age,
          y=over_50.chol,
          c=over_50.target)
ax.set(title='Heart diease(test)',
      xlabel='Age',
      ylabel='Cholestron')

ax.legend(*scatter.legend_elements(), title='Target')
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

scatter = ax0.scatter(x=over_50.age,
           y=over_50.chol,
           c=over_50.target)
ax0.set(title='Heart Disease and Cholestrol Level',
       ylabel='Cholestrol')
ax0.legend(*scatter.legend_elements(),
          title='Target')
ax0.axhline(y=over_50.chol.mean(), linestyle='--');

scatter = ax1.scatter(x=over_50.age,
                     y=over_50.thalach,
                     c=over_50.target)
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')
ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y=over_50.thalach.mean(),
           linestyle='--');

fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");
```
