# Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Preparing the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(15, 10)) # width and height

# Plot the data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from dictionary
gun_prices = {"Makarov": 150,
              "AK-47": 200,
              "Istiqlal": 500,
             "Colt": 250}
fig, ax = plt.subplots()

ax.bar(gun_prices.keys(), gun_prices.values())
# ax.plot(gun_prices.keys(), gun_prices.values())
# ax.scatter(gun_prices.keys(), gun_prices.values())

ax.set(title="Weapon Prices",
      xlabel="Name of Guns",
      ylabel="Price")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax1.plot(x, x-2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(gun_prices.keys(), gun_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# Subplot option 2
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(gun_prices.keys(), gun_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[$/./,]', '')
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("01/01/2021",
                                     periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales.Price.astype(int).cumsum()
car_sales
```

```python
# Let's plot
car_sales.plot(x="Sale Date", y="Total Sales")
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
