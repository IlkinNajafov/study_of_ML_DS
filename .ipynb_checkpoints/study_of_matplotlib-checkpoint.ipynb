# Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Preparing the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(15, 10)) # width and height

# Plot the data
ax.plot(x, y)

# Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Let's make ours

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [0, 15, 0]
y = [0, 15, 0]

fig, ax = plt.subplots(figsize=(15, 10))

plt.plot(x, y);
ax.set(title="This is Title",
      xlabel="This is X axis",
      ylabel="This is Y axis")

fig.savefig("images/another-plot.png")
```

## Making figures with NumPy arrays
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use same data to makke a Scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary
gun_prices = {"Makarov": 150,
              "AK-47": 200,
              "Istiqlal": 500,
             "Colt": 250}
fig, ax = plt.subplots()

ax.bar(gun_prices.keys(), gun_prices.values())
# ax.plot(gun_prices.keys(), gun_prices.values())
# ax.scatter(gun_prices.keys(), gun_prices.values())

ax.set(title="Weapon Prices",
      xlabel="Name of Guns",
      ylabel="Price")
```

```python
# let's try with horizontal bar
```

```python
fig, ax = plt.subplots()
ax.barh(list(gun_prices.keys()), list(gun_prices.values()));
```

```python
x = np.random.randn(1000)
x[:5]
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
ax1.plot(x, x-2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(gun_prices.keys(), gun_prices.values())
ax4.hist(np.random.randn(1000))
```
